```python
"""
 this code reads budget_data.csv
 calculate and print a few statistics
  * total months
  * total profit/losses
  * average changes month-to-month
  * maximum profit increases month-to-month
  * maximum profit decreases month-to-month
"""
# mount google drive to get file
from google.colab import drive
drive.mount("/content/drive")

import os
import csv

def read_csv(file_path):
  data = []
  with open(file_path,'r') as file:
    reader = csv.DictReader(file)
    for row in reader:
      data.append(dict(row))
  return data 

def get_change(profit_losses):
  changes = []
  for year,number in enumerate(profit_losses):
    if year==0:
      change = 0
    else:
      change = number - profit_losses[year-1]
    changes.append(change)
  return changes

def get_max_change(changes):
  max_change = max(changes)
  for idx in range(len(changes)):
    if changes[idx] == max_change:
      print(f"Greatest Increase in Profits: {dates[idx]} ${max_change}")
      return

def get_min_change(changes):
  min_change = min(changes)
  for idx in range(len(changes)):
    if changes[idx] == min_change:
      print(f"Greatest Decrease in Profits: {dates[idx]} ${min_change}")

# read budget_data.csv into list of dictionaries budget_data
current_dir = os.getcwd()
file_location = os.path.join(current_dir,"drive/MyDrive/git_repo/python-homework/resources","budget_data.csv")
budget_data = read_csv(file_location)

# desemble budget_data into two lists: dates, profit_losses
dates = []
profit_losses = []
for row in budget_data:
  dates.append(row['Date'])
  profit_losses.append(int(row['Profit/Losses']))

# calculate month-to-month profit/losses changes
changes = get_change(profit_losses)

print(f"Total Months: {len(budget_data)}")
print(f"Total: ${sum(int(row['Profit/Losses']) for row in budget_data)}")
print(f"Average   Change: ${round(sum(changes)/len(changes),2)}")
get_max_change(changes)
get_min_change(changes)
```
